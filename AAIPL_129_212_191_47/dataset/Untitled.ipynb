```python
import json
import random
import os

# Paths to the datasets
path_1 = "1.json"
path_2 = "2.json"
path_llama = "llama_70b.json"
path_truth_liar = "truth_and_liar.jsonl"
path_family = "family_reasoning_data_for_train.json"

# Function to load JSON file
def load_json(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)

# Function to load JSONL file
def load_jsonl(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    return data

# Load datasets
data_1 = load_json(path_1)
data_2 = load_json(path_2)
data_llama = load_json(path_llama)
data_truth_liar = load_jsonl(path_truth_liar)
data_family = load_json(path_family)

# Get all data from dataset/1.json and dataset/2.json
all_1 = data_1
all_2 = data_2

# Calculate how many points needed from llama_70b to reach 1000 for the first part
num_all_1 = len(all_1)
num_all_2 = len(all_2)
num_needed_llama = 1000 - num_all_1 - num_all_2

if num_needed_llama < 0:
    raise ValueError("Combined size of dataset/1.json and dataset/2.json exceeds 1000")

# Randomly sample from llama_70b if needed
if num_needed_llama > 0:
    if len(data_llama) < num_needed_llama:
        raise ValueError(f"llama_70b.json has fewer than {num_needed_llama} data points")
    llama_sample = random.sample(data_llama, num_needed_llama)
else:
    llama_sample = []

# Randomly sample 1000 points from truth_and_liar
if len(data_truth_liar) < 1000:
    raise ValueError("truth_and_liar.jsonl has fewer than 1000 data points")
truth_liar_sample = random.sample(data_truth_liar, 1000)

# Randomly sample 1000 points from family_reasoning_data
if len(data_family) < 1000:
    raise ValueError("family_reasoning_data_for_train.json has fewer than 1000 data points")
family_sample = random.sample(data_family, 1000)

# Add topic to family_reasoning_data samples
for data_point in family_sample:
    data_point['topic'] = "Puzzles involving generations and family tree logic"

# Combine all data
new_dataset = all_1 + all_2 + llama_sample + truth_liar_sample + family_sample

# Verify total size
if len(new_dataset) != 3000:
    raise ValueError(f"New dataset size is {len(new_dataset)}, expected 3000")

# Shuffle the combined dataset
random.shuffle(new_dataset)

# Save the new dataset
output_path = "new_dataset.json"
with open(output_path, 'w') as f:
    json.dump(new_dataset, f, indent=4)

print(f"New dataset created with {len(new_dataset)} data points and saved to {output_path}")
```

```python
import json

with open("/jupyter-tutorial/AAIPL_129_212_191_47/dataset/2.json", "r") as f:
    questions = json.load(f)

updated_questions = []
for q in questions:
    updated_q = {
        'topic': q['topic'],
        'question': q['question'],
        'choices': q.pop('choice'),
        'answer': q['answer'].split(')')[0],
        'explanation': q.get('explanation', '')
    }
    updated_questions.append(updated_q)
    
with open("input_eval2.json", "w") as f:
    json.dump(updated_questions, f, indent=4)
```

```python
with open("output_eval2.json", "r") as f:
    answers = json.load(f)
```

```python
correct = 0
for a, b in zip(updated_questions, answers):
    if a is None or b is None:
        continue

    # print(a.get("answer"),b.get("answer"))
    if a.get("answer") and b.get("answer") and a["answer"] == b["answer"]:
        correct += 1

accuracy = correct / len(answers)
print(f"Accuracy: {accuracy:.2f}")
```
